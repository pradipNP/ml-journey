# Handling the missing values (dropna & fillna)

```python
import pandas as pd

var = pd.read_csv("CSV_files\\8Worldwide_Protests.csv")
var
```

# 1. dropna() 
- It will delete all the data where there is NaN i.e. if found in any row or column, it will delete entire row or column

```python
var.dropna() # It will drop according to the row i.e. default axis=0
```

```python
var.dropna(axis=1) # drop by columns
```

### dropna(how = "any") :- It will also drop all NaN values containing row

```python
var.dropna(how = "any")
```

### dropna(how="all") :- It will drop only the row where there is all NaN value in row 

```python
var.dropna(how = "all")
```

### subset parameter :- column wise NaN value deleted

```python
var.dropna(subset=["Injuries"])
```

### inplace parameter :- Remove NaN value 

```python
var.dropna(inplace=True)
var
```

### thresh parameter :- drop according to your requirement

```python
var.dropna(thresh = 1)
var
```

# 2. fillna()
- This is used for replacing NaN values with else

```python
var1 = pd.read_csv("CSV_files\\8Worldwide_Protests.csv")
var1

print()
var1.fillna("Unknown")
```

### replace particular column wise names

```python
var1.fillna({"Year":"Not_remember",	"Country":"No_country",	"Protest_Name":"IDK", "Estimated_Participants":"Searching",	"Deaths":"Finding",	"Injuries":"NO_IDEA"})
```

### method parameter :- Used to copy same data of previous row or next row according to the forward and backward method

```python
var1.bfill() # used for backward copy

# or we can use,
# var1.fillna(method="ffill")
```

```python
var1.ffill() # used for forward copy
```

### column wise data filling
- using axis parameter

```python
var1.fillna(method="ffill", axis=1) #For forward filling
```

```python
var1.fillna(method="bfill", axis=1) # For backward filling
```

### inplace parameter :- if True it will replace NaN with the value we provided

```python
var1.fillna(111111, inplace=True)
var1
```

### limit parameter :- this will replace the NaN value with what we want but upto some limit column wise

```python
var = pd.read_csv("CSV_files\\8Worldwide_Protests.csv")
var
var.fillna("pkpkpk", limit=2)
```

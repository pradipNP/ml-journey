# Reshape the panda files
1. Melt Function
2. Pivot table

## 1. Melt()

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6], "eng":[10,12,14,15,16,12], "maths":[17,18,19,13,14,16]})
var

# Since it is showing horizontally, Let we organize the file vertically by reshaping it
```

```python
pd.melt(var)
```

```python
# Let's create ID of any specific column

pd.melt(var, id_vars=["days"])
```

```python
# Changing the columns name

pd.melt(var, id_vars=["days"], var_name="Subject", value_name="marks")
```

## 2. pivot()

```python
var
```

```python
var.insert(1, "st_name", ["a","b","c","a","b","c"])
var
```

```python
# To classify these columns like:
# creating a table where eng and maths score stores according to the students name day wise

var.pivot(index="days", columns="st_name")
```

```python
# To see only single subject data,
var.pivot(index="days", columns="st_name", values="eng")
```

- pivot_table() :- basically used for arithmetic/mathematical operations

```python
# To find mean
var.pivot_table(index="st_name", columns="days", aggfunc="mean")
```

```python
# To find sum
var.pivot_table(index="st_name", columns="days", aggfunc="sum")
```

- margin parameter :- Average value

```python
var.pivot_table(index="st_name", columns="days", aggfunc="mean", margins="True")
```

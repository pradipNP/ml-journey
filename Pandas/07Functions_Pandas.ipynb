# Change the name/value of any data in the dataframe 
- Method 1 (Shows error, not recommended)

```python
csv_1["Generations"][0] = "Generation_M"
csv_1
```

2. Method 2 (Recommended)
   - loc[row_no, column_name] = "replace_word"

```python
csv_1.loc[0, "Generations"] = "M"
csv_1
```

# To get a particular data

```python
# Let we want to find the row 2 and 3 and column Generations and Birth_Years

csv_1.loc[[2,3], ["Generations", "Birth_Years"]] # if we want all columns, then we can pass , :]
```

```python
# To get all data of these both above columns

csv_1.loc[:, ["Generations", "Birth_Years"]]
```

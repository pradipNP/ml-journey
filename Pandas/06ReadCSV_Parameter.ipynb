# Read CSV files

```python
import pandas as pd

csv_1 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv")
csv_1
```

```python
#Print data upto some rows

csv_2 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", nrows=5)
csv_2
```

```python
# To get a particular Column

csv_3 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", usecols=["Birth_Years","Generations"])
csv_3
```

```python

# Note: we can also pass the index no. of any column
csv_4 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", usecols=[0,4]) 
csv_4
```

### Skipping rows

```python

csv_5 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", skiprows=[1,10])
csv_5
```

## Index Column

```python
csv_6 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", index_col="Generations")
csv_6
```

## Changing Header

```python
csv_7 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", header = 1) # row 1 becomes new header
csv_7
```

```python
csv_8 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", names=["col","new"]) #names must be different
csv_8
```

## Remove Header

```python
csv_9 = pd.read_csv("CSV_files\\6Generations_M_to_Z.csv", header = None,)
csv_9
```

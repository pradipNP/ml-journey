# Handling Missing Values
1. Replace :- In Pandas, the replace() method is used to substitute existing values within a DataFrame or Series with new values.
2. Interpolate :- The interpolate() method in Pandas is used to fill missing values (NaN) in a DataFrame or Series using various interpolation techniques.

# 1. Replace

```python
import pandas as pd

var = pd.read_csv("CSV_files\\8Worldwide_Protests.csv")
var
```

## to_replace parameter :- used for what to replace with what 

```python
var.replace(to_replace="USA", value="United States")
```

```python
# without to_replace parameter

var.replace(["Nepal", "UK", "Greece"], "N")
```

## regex parameter :- regular expression

```python
var.replace("[A-Za-z]", "pk", regex=True)
```

## Replace using dictionary

```python
var.replace({"Country":"[A-Z]"}, 22, regex=True)
```

## inplace parameter -> directly replace in the original dataset
- var.replace(value_to_replace, method="ffill", inplace=True) ---> This will replace the value_to_replace with the ffill in the original dataset.

# 2. Interpolate
- The interpolate() method in Pandas is used to fill missing values (NaN) in a DataFrame or Series using various interpolation techniques.
- Note: If we are directly using Interplate function then it will work only with the number

```python
var
```

```python
var.interpolate() # It is showing error because it only work with number not with alphabet
# And it will linearly change the data
```

## method:{'linear','time','index','values','nearest','zero','slinear','quadratic','cubic','barycentric','krogh','polynomial','spline','piecewise_polynomial','from_derivatives','pchip','akima'}


```python
var.interpolate(method="index")
```

## axis parameter :- axis=0 works with columns wise

```python
var.interpolate(method="index", axis=0) # error because it works with numerical value
```

## limit parameter :- fills data only upto limit

```python
var.interpolate(limit=2)
```

## limit_direction :- used to fill data in the direction
- forward
- backward
- both

```python
var
var.interpolate(limit_direction="forward", limit=2) 
```

## limit_area parameter :- says either we want to use interpolate or not by specifying area
- limit_area = "inside"  ==> Fills the blank data
- limit_area = "outside" ==> Puts the NaN value

```python
var.interpolate(limit_area = "inside") # Fills the blank data
```

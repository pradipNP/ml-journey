# 1. Insert
- insert(index_no, position_of_new_column_&name, data_to_insert)

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4,5], "B":[9,8,7,6,5]})

var
```

```python
#Inserting same element of A column

var.insert(1,"Python", var["A"])
var
```

```python
#Inserting new data as a list
var.insert(1,"Python_1", [11,12,13,14,15])
var
```

```python
#Inserting new column with same elements of 'A' column upto index no. 3 by using slicing

var["new"] = var["A"][:3]
var
```

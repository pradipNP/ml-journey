# 1. Join

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]})
var2 = pd.DataFrame({"C":[1,2,3,4], "D":[11,12,13,14]})

var1.join(var2)
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]})
var4 = pd.DataFrame({"C":[1,2], "D":[11,12]})

var3.join(var4)
```

## index parameter
- To change the index name

```python
var5 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]}, index = ["a","b","c","d"])
var6 = pd.DataFrame({"C":[1,2], "D":[11,12]}, index = ["a","b"])

var5.join(var6)
```

```python
var5 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]}, index = ["a","b","c","d"])
var6 = pd.DataFrame({"C":[1,2], "D":[11,12]}, index = ["a","b"])

var6.join(var5)
```

## how parameter :- Used to print all elements 
- inner
- outer
- left
- right

```python
var5 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]}, index = ["a","b","c","d"])
var6 = pd.DataFrame({"C":[1,2], "D":[11,12]}, index = ["a","b"])

var6.join(var5, how = "right")
```

```python
var6.join(var5, how = "left")
```

```python
var6.join(var5, how = "inner") #shows common data
```

```python
var6.join(var5, how = "outer") # union wise
```

## If the column name is same:- suffix parameter is used
- lsuffix --> leftsuffix
- rsuffix --> rightsuffix

```python
var7 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]}, index = ["a","b","c","d"])
var8 = pd.DataFrame({"C":[1,2], "B":[11,12]}, index = ["a","b"])

var8.join(var7, how="outer", lsuffix="_12", rsuffix="_123")
```

# 2. Append
- In the join function, if there is the same column name, we use suffix parameter
- In the append function, if there is the same column name or different, it works same for both

```python
var_1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]})
var_2= pd.DataFrame({"C":[1,2], "D":[11,12]})

var_1.append(var_2) # pd.concat([var_1, var_2]) is the modern replacement for df1.append(df2)
```

## Note: pd.concat([var_1, var_2]) is the modern replacement for df1.append(df2)

```python
var_1 = pd.DataFrame({"A":[1,2,3,4], "B":[11,12,13,14]})
var_2= pd.DataFrame({"C":[1,2], "D":[11,12]})

pd.concat([var_1, var_2]) # It will show duplicate index
```

## To reset the indices,
- ignore_index parameter is used

```python
pd.concat([var_1, var_2], ignore_index=True)
```
